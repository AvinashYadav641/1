# Collaborative Filtering Based recomender System

```python
ratings_with_name.head(2)
```

```python
x =ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
true_readers = x[x].index
```

```python
# Boolean series
ratings_with_name['User-ID'].isin(true_readers)
```

```python
# Now we insert this series in the ratings with name
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(true_readers)]
filtered_rating.head(2)
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books  = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
# We are filling 0 in Cans of NA
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python

sorted(list(enumerate(similarity_scores[0])), key=lambda x:x[1], reverse=True)[1:6]
# here key=lambda x:x[1] sort the list on the basis of similarity scores
```

```python
# we are creating a function will give us the recomende movie\

def recommend(book_name):
#     index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:11]
    
    data =[ ]
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Notebook')
```

```python
recommend('The Da Vinci Code')
```

```python
recommend('The Fellowship of the Ring (The Lord of the Rings, Part 1)')
```
